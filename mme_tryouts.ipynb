```python
import pandas as pd

# create a dataframe with only the ExtendedSessionID column from the 71M SharedResponses.csv file
reader = pd.read_csv('SharedResponses.csv', usecols=['ExtendedSessionID'])
```

```python
# use this dataframe to create a dataframe (reader_24) that contains only the ExtendedSessionID rows that appear 24 times

session_counts = reader['ExtendedSessionID'].value_counts()

ids_to_keep = session_counts[session_counts == 24].index

reader_24 = reader[reader['ExtendedSessionID'].isin(ids_to_keep)]
```

```python
# print the shape of the resulting dataframe
print(reader_24.shape)

# there are 17756736 rows. 17756736/24 = 739864 unique ExtendedSessionIDs.
```

```python
# to make it more computationally efficient, only keeping the unique ExtendedSessionIDs (so only 1 time in the list, instead of 24 times for each ExtendedSessionID)

reader_24_unique = reader_24['ExtendedSessionID'].unique()
```

```python
print(len(reader_24_unique))    # should be 739864
print(reader_24_unique)         # to check to see if it contains only unique ExtendedSessionIDs
```

```python
# transform to a list to feed to the for loop that will extract all the corresponding columns from the SharedResponses.csv file

reader_24_list = reader_24_unique.tolist()
```

```python
len(reader_24_list)     # should be 739864
```

```python
# empty dataframe to append the rows of the sessions with 24 rows to

subset = pd.DataFrame()
```

```python
# reading SharedResponses to extract the rows with ExtendedSessionID in reader_24_list
# this will result in a dataframe 'subset' that contains around 17 million rows (739864 * 24)
# these are then all sessions with 12 scenarios (24 rows)


chunk_size = 100_000
reader = pd.read_csv('SharedResponses.csv', chunksize=chunk_size, dtype=str, low_memory=False)

for i, chunk in enumerate(reader):
    
    print(f"Processing chunk {i+1}")

    # Filter rows where ExtendedSessionID is in reader_subset
    subset_chunk = chunk[chunk['ExtendedSessionID'].isin(reader_24_list)]

    # Append filtered chunk to empty df
    subset = pd.concat([subset, subset_chunk], ignore_index=True)

    print(f"Finished processing chunk {i+1}")

print("All chunks have been processed and combined.")
```
