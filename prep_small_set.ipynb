```python
# what number of rows that is from western countries?

total_western = df_total_complete['UserCountry3'].isin(countries).sum()
print(f"Number of rows Western: {total_western}")

# this is % of the dataset that is from Western countries
print(f"Rounded % Western: {round(total_western /len(df_total_complete)*100)}")    # rounded percentage
print(f"Unrounded % Western: {total_western /len(df_total_complete)*100}")           # exact percentage
```

```python
# dropping the columns that are not needed

df_total = df_total_complete.drop(['ScenarioType', 'ScenarioOrder', 'DescriptionShown', 'LeftHand', 'UserCountry3'], axis=1)
```

```python
df_total.shape
```

```python
dupli = df_total.duplicated().sum()
```

```python
print(f"Number of duplicates: {dupli}")
```

```python
# checking the missing values

df_total.isnull().sum().sum()

# there are no missing values in the dataset
```

```python
# binarizing UserID
# we know from EDA of llm_dataset that userID for llms are always 'chatbot'. We will encode this as 1 and all other userIDs as 0 (which are humans)

df_total['UserID'] = df_total['UserID'].apply(lambda x: 1 if x == 'chatbot' else 0)
```

```python
df_total['UserID'].value_counts()
```

```python
# checking if 2% is llm
print(df_total['UserID'].value_counts(normalize=True))

# yes it is true
```

```python
# randomly shuffling the dataset so that not all llm's are on top of the dataset

df_total = df_total.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df_total.head(10)
```

```python
df_total.info()
```

```python
# Get the maximum float value across the entire DataFrame (ignoring non-numeric columns)
max_value = df_total.select_dtypes(include=['float', 'int']).max().max()

print(f'Highest number in the dataset: {max_value}')

# can do int8 cause the max value is 5
```

```python
# change datatypes of the columns

# Convert all float64 and int64 columns to int8
df_total = df_total.astype({col: 'int8' for col in df_total.select_dtypes(include=['float64', 'int64']).columns})

# Convert all object columns to category
df_total = df_total.astype({col: 'category' for col in df_total.select_dtypes(include=['object']).columns})
```

```python
df_total.info()
```

```python
print(len(df_total))
print(len(df_total)/100*2)
print(len(df_total)/100*98)
```

```python
# saving the dataset
df_total.to_csv('sub_dataset.csv', index=False)
```
