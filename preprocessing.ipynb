## Mme total & llm's

### Creating 98 dataset

```python
# reading datasets

import pandas as pd

df_mme = pd.read_csv('mme_98_dataset.csv')
```

```python
df_mme.shape
```

```python
df_llm = pd.read_csv('llm_125_dataset.csv')
```

```python
df_llm.shape
```

```python
# randomly keeping 2.5M - 2.45M of the dataset = 50.000 rows of LLM needed
# 50000 rows = 50000 / 2 = 25000 unique responseID's
```

```python
# randomly keep 25000 unique responseID's LLM dataset

# Getting unique UserIDs
Response_unique = df_llm['ResponseID'].unique()
print(len(Response_unique))  # should be 1547132/2 = 773566

# Selecting 25000 UserIDs from the unique set
Response_keep_llm = pd.Series(Response_unique).sample(n=25000, random_state=42)
print(len(Response_keep_llm))  # should be 25000
```

```python
df_llm.head()
```

```python
df_llm_sub = df_llm[df_llm['ResponseID'].isin(Response_keep_llm)]
```

```python
df_llm_sub.shape

# should be 50000 rows
```

```python
# concatanating the datasets, with mme after llm cause of ResponseID, and resetting the index

df_concat = pd.concat([df_llm_sub, df_mme], axis=0)
df_total_complete = df_concat.reset_index(drop=True)
```

```python
df_total_complete.shape # should be 2.5M and 31 cols
```

```python
df_total_complete.duplicated().sum()
```

```python
df_total_complete.head()
```

```python
df_total_complete.tail()
```

```python
df_total_complete.columns
```

```python
# check distribution of the target variable (UserID)

df_total_complete['UserID'].value_counts(normalize=True)
```

```python
# showing all the columns
pd.set_option('display.max_columns', None)
```

```python
df_total_complete['ResponseID'].value_counts().min()
```

```python
df_total_complete['ResponseID'].value_counts().max()
```

```python
dupli = df_total_complete.duplicated().sum()
```

```python
print(f"Number of duplicates: {dupli}")
```

```python
# checking the missing values

df_total_complete.isnull().sum().sum()

# there are no missing values in the dataset
```

```python
# randomly shuffling the dataset so that not all llm's are on top of the dataset

df_total = df_total_complete.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df_total.head()
```

```python
df_total.info()
```

```python
# Get the maximum float value across the entire DataFrame (ignoring non-numeric columns)
max_value = df_total.select_dtypes(include=['float', 'int']).max().max()

print(f'Highest number in the dataset: {max_value}')

# can do int8 cause the max value is 5
```

```python
# change datatypes of the columns

# Convert all float64 and int64 columns to int8
df_total = df_total.astype({col: 'int8' for col in df_total.select_dtypes(include=['float64', 'int64']).columns})

# Convert all object columns to category
df_total = df_total.astype({col: 'category' for col in df_total.select_dtypes(include=['object']).columns})
```

```python
df_total.info()
```

```python
# resetting index
df_total = df_total.reset_index(drop=True)
```

```python
df_total.head()
```

```python
# saving the dataset
df_total.to_csv('total_98_dataset.csv', index=False)
```
